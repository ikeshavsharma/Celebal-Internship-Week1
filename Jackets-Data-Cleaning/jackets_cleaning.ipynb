# Handling Missing Values

```python
df['discount'] = df['discount'].fillna(0)

df['what_customers_said'] = df['what_customers_said'].fillna("No Review")

df['seller_name'] = df['seller_name'].fillna("Unknown Seller")

df['videos'] = df['videos'].fillna("No Video")

df['seller_information'] = df['seller_information'].fillna("No Information")

df['variations'] = df['variations'].fillna("No Variation")
```

# Verifying Missing Values Again

```python
df.isnull().sum()
```

# Checking Duplicate Rows

```python
df.duplicated().sum()
```

# Removing Duplicate Rows

```python
df = df.drop_duplicates()
```

# Selecting Important Columns

```python
df[['title', 'rating', 'final_price']].head()
```

# Filtering Expensive Jackets

```python
expensive_jackets = df[df['initial_price'] > 5000]

expensive_jackets[['title', 'initial_price', 'final_price']]
```

# Cleaning final_price Column

```python
df['final_price'] = df['final_price'].replace('[₹,"]', '', regex=True)

df['final_price'] = df['final_price'].astype(float)
```

# Creating Derived Column

```python
df['price_difference'] = df['initial_price'] - df['final_price']
```

# Displaying Derived Column

```python
df[['initial_price', 'final_price', 'price_difference']].tail()
```

# Saving Cleaned Dataset

```python
df.to_csv("cleaned_jackets.csv", index=False)

print("Saved successfully!")
```
